## Time sync
Time sync is needed because Google Colab's clock is sometimes wrong.

### Why it matters:
- Many websites (especially Hugging Face) use SSL   certificates that check the current time.
- If Colab's time is even a few minutes off, the - certificate check fails.
- Then you cannot download the Echo-TTS models → you get errors or stuck downloads.

```python
# 4. Enable Gradio share=True
from pathlib import Path
file_path = Path("gradio_app.py")
text = file_path.read_text(encoding="utf-8")

new_launch = '''if __name__ == "__main__":
    demo.launch(
        share=True,
        server_name="0.0.0.0",
        server_port=7860,
        allowed_paths=[str(AUDIO_PROMPT_FOLDER)]
    )'''

if "if __name__ == \"__main__\"" in text:
    text = text.replace(
        '''if __name__ == "__main__":
    demo.launch(
        allowed_paths=[str(AUDIO_PROMPT_FOLDER)]
    )''',
        new_launch
    )
    file_path.write_text(text)
    print("✅ gradio_app.py updated with share=True")
```

```python
# 5. Launch Echo-TTS
print("⏳ Starting Echo-TTS... (first run downloads models ~1-2 min)")
!python gradio_app.py
```

Sometime Gradio Public Link was not Work. So you can use NGROK tunnel.

```python
# Install & run Ngrok - Optional
!pip install pyngrok -q

from pyngrok import ngrok, conf
import getpass

print("\n" + "="*60)
print("🚀 PASTE YOUR NGROK AUTHTOKEN BELOW")
print("="*60)

token = getpass.getpass("ngrok authtoken: ")
conf.get_default().auth_token = token

tunnel = ngrok.connect(7860, "http")
print(f"\n✅ YOUR PUBLIC LINK:\n{tunnel.public_url}")
print("   (Wait 10-20 seconds after the app starts)\n")

# Launch Echo-TTS
print("⏳ Starting Echo-TTS... (first run downloads models ~1-2 min)")
!python gradio_app.py
```
